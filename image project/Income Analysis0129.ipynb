```python
# import libraries
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('h1b_datahubexport-2020.csv',skiprows=1)
data
```

```python
endata= data[data['NAICS']==81]
totala = sum(endata['Initial Approval']) + sum(endata['Initial Denial']) + sum(endata['Continuing Approval']) + sum(endata['Continuing Denial'])
print('H-1B total Tech applicant numbers: %s' %totala)
print('H-1B Tech applicants initial approval: %s' %sum(endata['Initial Approval']))
print('H-1B Tech applicants Continuing Approval: %s' %sum(endata['Continuing Approval']))
print('H-1B Tech applicants initial approval ratio: %%%s' % ( sum(endata['Initial Approval']) / totala * 100 ))
print('H-1B Tech applicants Continuing Approval ratio: %%%s' % ( sum(endata['Continuing Approval'])/ totala * 100 ))
```

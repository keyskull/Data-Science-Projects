```python
from selenium import webdriver
from time import sleep
import numpy as np
import pandas as pd
from pattern.web import URL
import os, io, requests
from PIL import Image 
```

```python
driver = webdriver.Chrome()
driver.get('https://tinder.com')
```

```python
def dislike():
    dislike_btn = driver.find_element_by_xpath('//main/div/div/div/div/div[2]/div/div/div[2]/button')
    dislike_btn.click()
```

```python
def save_image(path):
    url = URL(path)
    uri = 'res/' + url.domain + '/' + '/'.join(url.path)
    if not os.path.exists(uri):
        os.makedirs(uri)
    Image.open(requests.get(path, stream=True).raw).save(uri + '/'+ url.page)
```

```python
def get_persent_data():
    try:
        info_button = driver.find_element_by_xpath('//main/div/div/div/div/div/div[1]/div[3]/div[3]/button')            
        info_button.click()
    except Exception as inst:
        print(inst)
    ids = ''
    age = 0
    name = ''
    addition_info = []
    images_url = []
    description = ''
    try:
        ids = driver.find_element_by_xpath('//main/div/div/div/div/div[1]/div/div[1]/span/div/div[1]/span/div/div').value_of_css_property('background-image').split('/')[3]
        detail = driver.find_element_by_xpath('//main/div/div/div/div/div[1]/div/div[2]/div[1]/div/div[1]')
        name = detail.find_element_by_xpath('div[1]/h1').text
        age = detail.find_element_by_xpath('span').text
        addition_info = driver.find_element_by_xpath('//main/div/div/div/div/div[1]/div/div[2]/div[1]/div/div[2]')
        addition_info = [x.text for x in addition_info.find_elements_by_xpath('div/div')]
        try:
            description = driver.find_element_by_xpath('//main/div/div/div/div/div[1]/div/div[2]/div[2]')
            description = description.text
        except:
            print("No Description")
            
        images_row = driver.find_elements_by_xpath('//main/div/div/div/div/div[1]/div/div[1]/span/div/div[2]/button')
        
        for x in images_row:
            sleep(0.1)
            url = driver.find_element_by_xpath('//main/div/div/div/div/div[1]/div/div[1]/span/div/div[1]/span[@aria-hidden= "false"]/div/div').value_of_css_property('background-image').split('"')[1]
            images_url.append(url)
            save_image(url)
            x.click()
    except Exception as inst:
        print(inst)
        sleep(2)
        return get_persent_data()

    return pd.DataFrame(data={'id':ids,'name': name,'age':age,'addition_info': [np.array(addition_info)], 'description':description,'images_url': [images_url]},index=[0])
```

```python
# data = pd.DataFrame(columns=['id','name','age','addition_info','description','images_url'])
data = pd.read_csv('res/data.csv')
```

```python
for x in range(5000):
    data = data.append(get_persent_data(),ignore_index=True)
    data.to_csv('res/data.csv', encoding= 'utf-8-sig')
    dislike()
    sleep(1)
```
